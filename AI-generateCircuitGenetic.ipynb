## <div style="direction:rtl"> فاز سوم: پیاده سازی و مشخص کردن معیار سازگاری</div>
<div style="direction:rtl"> عاملی که باعث می شود ما در الگوریتم بتوانیم پیش برویم و بین کروموزوم ها با سنجه ای عددی تفاوت قائل شویم یک معیار سازگاری دقیق است. در اینجا معیار سازگاری برای هر کروموزوم تعداد سطر هایی از فایل داده شده است که به ازای آن ورودی ها خروجی خواسته شده را تولید کند. بدین شکل کروموزوم هایی که با محیط سازگارترند بدین معنی که منطبق با سطر های بیشتری هستند معیار سازگاری بیشتری هستند. برای پیاده سازی این قسمت توابع و قسمت های مختلفی پیاده سازی شد که در ادامه به آنها اشاره خواهیم کرد:<br>
    - تعریف خروجی های هر گیت:<br>
    برای این قدم تابع زیر را تعریف شده که یک دیکشنری دارد با کلید گیت ها و خروجی متناسب با هر کدام.</div>

```python
def getEachResult(self,input1, input2, operator):
        operation = {
            "AND": (input1 and input2),
            "NAND": (not(input1 and input2)),
            "OR": (input1 or input2),
            "NOR": (not(input1 or input2)),
            "XOR": (input1 ^ input2),
            "XNOR": (not(input1 ^ input2))
        }
        return operation.get(operator)
```

<div style="direction:rtl">
    - تعریف محاسبه خروجی مجموعه ورودی ها به ازای مجموعه از گیت ها:<br>
    بدین جهت تابعی تعریف کردیم که مجموعه ورودی ها و مجموعه گیت ها را بگیرد و با کمک آنها مجموعه از خروجی ها را نیز تولید کند و با کمک عباراتی نظیر به نظیر و با ترتیبی خاص خروجی کلی را تولید کند.</div>

```python
def getResultedOutput(self, row, operators):
        inputs = self.getInputs(row)
        outputs = []
        for i in range(len(operators)):
            if i == 0:
                outputs.append(self.getEachResult(inputs[i], inputs[i+1], operators[i]))
            else:
                outputs.append(self.getEachResult(outputs[i-1], inputs[i+1], operators[i]))
        return outputs[-1]
```

<div style="direction:rtl">
    - محاسبه کلی معیار سازگاری:<br>
    بدین جهت تابع زیر را تعریف کردیم که برای هر سطر ابتدا با کمک تابع بالا خروجی متناسب با گیت های داده شده را به دست می آورد. سپس با خروجی نوشته شده در سطر مقایسه می کند و در صورت برابری یک واحد به سازگاری این کروموزوم اضافه می کند.</div>

## <div style="direction:rtl"> فاز چهارم: پیاده سازی mutation و crossover و تولید جمعیت بعدی </div>
<div style="direction:rtl"> حال برای رسیدن به جواب باید نسل بعدی را تولید کنیم و همانطور تا زمانی که به کروموزومی با معیار سازگاری برابر با تعداد سطرهای فایل ورودی برسیم(بدون در نظرگرفتن سطر اوّل که نام ستون هاست)ادامه دهیم. حال مهم ترین فاکتور در سرعت رسیدن به پاسخ تولید مناسب نسل بعدی از این نسل است. بدین جهت ابتدا یک  انتخاب از نسلی که داریم باید انجام شود که در فاز پنجم توضیح داده خواهد شد. سپس باید روی کروموزوم های انتخاب شده عملیات crossover و mutation را پیاده سازی کنیم که در ادامه هر یک را توضیح می دهیم:<br>
   -crossover:
    </div>

```python
def crossover(self, other):
        probability = random.uniform(0, 1)
        if probability <= 0.25:
            return (self, other)
        else:
            point1 = int(random.uniform(0, targetLen/2))
            point2 = int(random.uniform((targetLen/2)+1, targetLen-2))
            new1 = self.chromosome[0:point1] + other.chromosome[point1:point2] + self.chromosome[point2:]
            new2 = other.chromosome[0:point1] + self.chromosome[point1:point2] + other.chromosome[point2:]
            return (Chromosome(new1), Chromosome(new2))
```

<div style="direction:rtl"> ابتدا عددی را بین صفر و یک انتخاب می کنیم و اگر کمتر از بیست و پنج صدم بود همان دو کروموزوم اولیه را باز می گردانیم. بدین معنا که با احتمال بیست و پنج درصد کروموزوم ها را تغییر نمی دهیم. اما به احتمال هفتاد و پنج درصد دو کروموزوم کنار هم را به شکل زیر تغییر می دهیم:<br>
    دیدگاه کلی crossover از دو نقطه است. بدین شکل که دو نقطه را به صورت تصادفی انتخاب می کنیم. بعد و قبل این نقاط را تغییر نمیدهیم و بین آنها را با همان قسمت از کروموزوم دیگر تکمیل میکنیم.<br>
    -mutation:</div>

```python
def mutation(self, size):
        probability = 1 - (self.fitness/size)
        newChromosome = []
        for gene in self.chromosome:
            chance = random.uniform(0, 1)
            if chance <= probability:
                gene = self.getRandomGene()
            newChromosome.append(gene)
        return Chromosome(newChromosome)
```

### <div style="direction:rtl"> سوال چهارم:</div>
<div style="direction:rtl">در مقطعی از اجرای الگوریتم ممکن است کروموزوم ها و معیار سازگاری های یکسانی را در چند نسل متفاوت ببینیم و دلیل آن تصمیم گیری ها در انتخاب والد و عملیات crossover و mutation است. ممکن است در چند نسل والد ها از کروموزوم ها با سازگاری بالاتر انتخاب شود و با crossover هم تنها از دو کروموزوم خوب قسمت هایی را جدا کنیم و ترکیب کنیم که ساختاری مشابه حالت قبلی ایجاد کند که تفاوت چندانی در معیار سازگاری ایجاد نکند و همچنین با احتمالی که برای mutation در نظرگرفته ایم چون معیار سازگاری بالایی دارد با احتمال بسیار کمی تغییر می کند و کروموزوم یکسان به ما می دهد.از راه حل های این مشکل می تواند به موارد زیر اشاره کرد: <br>
    - تغییر معیار سازگاری به گونه ای که یک دسته از کروموزوم ها را ناگهان بی ارزش نکند.<br>
    - تغییر روش انتخاب به گونه ای که هنگام گیر افتادن بتواند از کروموزوم های کم ارزش تر با احتمال بالاتر انتخاب کند تا با ترکیب آنها گستردگی به الگوریتم برگردد.<br>
    - تغییر crossover به گونه ای که هنگام گیر افتادن با احتمال کمتری همان والد ها را به عنوان فرزندان به نسل بعد انتقال دهد.<br>
    - تغییر mutation به گونه ای که هنگامی که گیر افتادیم با احتمال بالاتری حتی برای کروموزوم های خوب تغییر ایجاد کنیم.<br>
    راه چهارم از آنجایی که بر روی یک الگوریتم تاثیر مستقیم دارد و همچنین برای کروموزم های خوب که عامل گیر افتادن ما هستند احتمال تغییر را بالا می برد به عنوان روش مناسب انتخاب و پیاده سازی شد.<br>
    بدین روش که بهترین معیار در چند نسل بررسی می شود و در صورت تکرار در چند مرحله یک flag ای فعال می شود و در صورت فعال بودن آن mutation در دور بعد مقداری را به احتمال تغییر اضافه می کند تا کروموزوم بهتر هم شانس بالاتری برای تغییر داشته باشد. این اعداد ثبت یعنی تعداد نسل های تکراری و مقدار افزایش احتمال هایپرپارامتر هستند که نتیجه بررسی آنها در آینده بیان می شود.</div> 

## <div style="direction:rtl">توضیحی کلی بر پیاده سازی </div>
<div style="direction:rtl"> در پیاده سازی سعی شده است با کمک گرفتن از پیاده سازی شی گرا نظم و در پایان رفع مشکل در کد را ساده تر کنیم. سه کلاس اصلی داریم:<br>
    -Chromosome:<br>
   در این کلاس توابعی مورد نیاز برای مدیریت کروموزوم ها پیاده سازی شده اند. از جمله ایجاد رندوم ژن ها و در نتیجه ساخت کروموزوم ها و محاسبه سازگاری و crossover و mutation.<br>
   -Generation:<br> 
   این کلاس عملاً یک لیست با ساختار heapq است که برای نگه داری کروموزوم های هر نسل استفاده می شود. توابع مربوط به نسل ها مانند انتخاب از نسل ها در آن آمده است.<br>
    -Test:<br>
   این کلاس برای مدیریت و کار با فایلی است که به عنوان تست داده شده است پیاده سازی شده است.توابع مدیریت ورودی مانند تشخیص ورودی ها و خروجی های هر مدار و محاسبه خروجی با گیت ها در این کلاس وجود دارد.<br>
   علاوه بر این سه کلاس دو متغیر global داریم:<br>
    - PossibleGenes:<br>
    ژن هایی که می توان در ساخت کروموزوم از آنها استفاده کرد.<br>
    -population<br>
    جمعیت هر نسل.<br>
    همچنین یک تابع main داریم که شکل کلی الگوریتم ژنتیک در آن پیاده شده است. به ترتیب نسل اولیه را می سازد سپس چک میکند به هدف رسیدیم یا نه. انتخاب را انجام می دهیم. بعد crossover و بعد mutation را انجام می دهیم و نسل جدید را می سازیم. تا به هدف برسیم ادامه می دهیم.<br>
نکته دیگری که به آن باید توجه شود آن است که مقادیر داده شده در ورودی string هستند و قبل از هر کاری آن ها را به boolean باید تبدیل کنیم و این کار با مقایسه string با "TRUE" انجام می شود.
</div>

## <div style="direction:rtl">Hyper parameter </div>
<div style="direction:rtl"> این پارامتر ها پارامترهایی هستند که با تغییر آنها سرعت یادگیری تغییر می کند. تعیین اینکه به طور دقیق کدام مقدار برای این پارامتر در مسئله بهتر است ساده نیست. چون تعداد زیادی متغیر با مقدار رندوم داریم که اگر آن پارامتر را تغییر دهیم و دوباره برنامه را اجرا کنیم، هیچ تضمینی وجود ندارد که مقادیر رندوم ثابت بمانند و بتوانیم مقایسه دقیقی داشته باشیم. حال این پارامتر ها را در این پروژه بیان میکنیم و مروری بر مقداری که به آنها داده ایم و دلیل آن انجام میدهیم:<br>
    -جمعیت:<br>
    جمعیتی که برای نسل ها در نظر میگیریم بسیار مهم است چون اگر از یک حدی بیشتر باشد تنها زمان اجرا را زیاد می کند بدون تاثیر در سرعت رسیدن به هدف و اگر از یک حدی کمتر باشد گستردگی زیادی در کروموزوم های نسل ها وجود ندارد و در نسل های دیرتری به جواب خواهیم رسید. بنابراین باید حالتی میانه را در نظر گیریم. در ادامه تعدادی نتیجه مربوط به جمعیت های متفاوت آمده است. تاگید می شود الزاماً خوب بودن نتیجه در یکی از این حالات به معنای خوب یودن همیشگی نیست به دلیل وجود تعداد زیادی متغیر تصادفی.<br></div>

<div style="direction:rtl">
    - احتمال crossover<br>
    همانطور که بالاتر هم گفته شد ما احتمال را به صورت بیست و پنج درصد بدون تغییر و هفتاد و پنج درصد تغییر در نظر گرفتیم. باید احتمال تغییر بالاتر باشد تا از رو کروموزوم های نسل قبل با تغییر کروموزوم های جدیدی ایجاد شود و گستردگی در نسل ها به وجود آید. البته بیش از حد بالا بردن این احتمال مناسب نیست زیرا کروموزوم هایی که خوب هستند را به احتمال بیشتری ترکیب می کنیم و ممکن است سازگاری آنها را کاهش دهیم.</div>

<div style="direction:rtl">
    - احتمال mutation<br>
    همانطور که گفته شد این احتمال به گونه ای پیاده سازی شده که زمانی که به یک کروموزوم با سازگاری بالا رسیدیم هر ژن احتمال کمتری برای تغییر داشته باشد. در کل این احتمال نباید زیاد باشد چون باعث خراب شدن کروموزوم سازگارتر می شود. امّا این کم بودن ممکن است باعث گیر افتادن در یک حالت شویم. که برای آن راه حلی ارائه شد و همان راه حل دو پارامتر دیگر ایجاد کرد:<br>
    × تعداد بار های تکرار تا بگوییم گیر افتادیم:<br>
    بدین معنا که بعد از چند دور که معیار ها تغییر نکردند بگوییم گیر افتادیم و تابع جهش را تغییر دهیم. ما بعد از دوبار را در نظر گرفتیم. هر چه زودتر این گیر افتادن مشخص شود بهتر است.<br>
    × مقدار افزایش احتمال تغییر:<br>
    بدین معنا که اگر در یک حالتی با سازگاری بالا گیر افتادیم احتمال جهش را به چه اندازه تغییر دهیم. ما تغییر به اندازه 0.1 را در نظر گرفتیم. نباید مقدار تغییر چنان زیاد باشد که تعداد زیادی از ژن ها تغییر کنند و از آن سازگاری خوب دور شویم.<br>
    در ادامه دو حالت را نشان دادیم اعداد روبروی A نشان دهنده مقدار سازگاری کروموزوم است.</div>

## <div style="direction:rtl">نتیجه گیری </div>
<div style="direction:rtl"> الگوریتم ژنتیک در کل برای حل مسائل با فضای حالت بزرگ و گسترده یا ناشناس بسیار مناسب است امّا میزان تناسب و قدرت آن بسیار وابسته به نوع پیاده سازی است. انتخاب روش مناسب برای محاسبه میزان سازگاری از اهمیت بالایی برخوردار است. همچنین انتخاب روش صحیح crossover و mutation هم حائز اهمیت است از آنجا که این موارد باعث تغییر کروموزوم ها هستند. تغییر مناسب و خوب می تواند ما را سریعتر به جواب برساند، تغییر غیر مناسب ما را در یک حالت بهینه محلی گیر می اندازد و حتی از جواب دور می کند. انتخاب جمعیت مناسب و به دور از جهت گیری هم از دیگر نکات مهم است. باید سعی شود به هیچ عنوان نسل ها را به یک سمت خاص هدایت نکنیم تا از حالت ها خوب ممکن دیگر دور نشویم. </div>
